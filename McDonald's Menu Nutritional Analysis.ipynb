#### 1.Now let us understand the data in depth using Exploratory Data Analysis (EDA).

Lets visualize the distribution of calorie count across menu items

```python
# Analyze the distribution of calorie counts across menu items
plt.figure(figsize=(10,6))
sns.histplot(df['Calories'],bins=20,kde=True,color='Purple')
plt.title('Distribution of Calorie Counts')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()
```

#### Analysis
 


####    The EDA shows that the distribution of Calorie count across menu items are right-skewed with most items falls less than 1000 calories.However there are some outliers with high calorie value.

#### 2. Explore nutritional content of different items

```python
# To Generate Random color 
import random

#Define a set of seaborn color palettes
seaborn_color = ['deep','muted','bright','pastel','dark','colorblind']

#Randomly select a color from the list
random_color = random.choice(seaborn_color)

# Explore nutritional content of different items
nutritional_content=['Total Fat','Saturated Fat','Cholesterol','Sodium','Carbohydrates','Dietary Fiber','Sugars','Protein']
plt.figure(figsize=(15,10))
for i,nutrient in enumerate(nutritional_content,1):
    plt.subplot(3,3,i)
    sns.histplot(df[nutrient],bins=20,kde=True,color=random.choice(sns.color_palette(random_color)))
    plt.title("Distribution of {}".format(nutrient))
    plt.xlabel(nutrient)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

####  Analysis   

#### While analysing the nutritional content of different items,we observe that the distribution vary across nutrients.
* The distribution of Total Fat and Cholesterol is right skewed,while the distribution of Carbohydrates is more evenly spread.

* The distribution of saturated fat and sugar suggests that it is maintaining a relatively constant value.
* Dietary Fiber need to increase in menu items where as Sodium need to decrease.
* Upon observing the chart, it is evident that the majority of menu items have protein content below 20, indicating a need for increased protein levels.

#### 3. Now lets find the trends and patterns in the nutritional data set

```python
# Drop the non-numeric columns before calculating the correlation matrix
numeric_df=df.drop(columns=['Category','Item','Serving Size'])
# Identify the correlation
correlation_matrix= numeric_df.corr()
plt.figure(figsize=(12,8))
sns.heatmap(correlation_matrix,annot=True,cmap='viridis',fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

#### Analysis  

#### The correlation matrix indicates that some nutrients have strong positive correlations, such as total fat and  saturated fat, which is expected.


#### 4.Data Visualization

Lets create Box plot to visualize calorie distribution and Category

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='Category',y='Calories',data=df)
plt.title('Calorie Distribution by Category ')
plt.xticks(rotation=45)
plt.ylabel('Calories')
plt.show()
```

#### Analysis

#### In terms of data visualization, we created box plots to visualize the distribution of calorie counts across different menu categories.

* From the plot we observe that Smoothies & Shakes have large calorie count when compared to other categories.
* Beverages also exhibit low calorie counts.

```python
# Bar charts to visualize the nutritional content
nutritional_content = ['Total Fat', 'Saturated Fat', 'Cholesterol', 'Sodium', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein']

plt.figure(figsize=(20, 15))
for i, nutrient in enumerate(nutritional_content, 1):
    plt.subplot(3, 3, i)
    sns.barplot(x='Category', y=nutrient, data=df)
    plt.title('Average {} by Category'.format(nutrient))
    plt.xlabel('Category')
    plt.ylabel(nutrient)
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### Analysis

#### Additionally, we used bar plots to compare the different nutritional content across categories.

* From the graph we can interpret that Breakfast have high Total Fat,Cholestrol,Saturated Fat.
* Sodium is high for chicken & fish
* Sugar and carbohydrate high for smoothies & Shakes
* Protein is high in Chicken and Fish

#### 5.Nutrition Based Insight

```python
# Lets find the Item with highest calories and lowest calorie

highest_calorie = df[df['Calories']==df['Calories'].max()]
lowest_calorie = df[df['Calories']==df['Calories'].min()]

print("Items with highest Calories Count : ")

# Transpose the DataFrame
highest_calories_transposed = highest_calorie.transpose()
print(highest_calories_transposed)


print("\n Items with lowest Calories Count : ")

# Transpose the DataFrame
lowest_calories_transposed = lowest_calorie.transpose()
print(lowest_calories_transposed)

```

#### Analysis

#### Through analysis we found that category 'Chicken & Fish' have highest Calorie count and Beverages,Coffee & Tea have lowest calorie Count

#### 6.Average Nutritional content of Popular menu categories

```python
# Determine the average nutritional content of popular menu categories
nutritional_content_1=['Total Fat','Saturated Fat','Cholesterol','Sodium','Carbohydrates','Sugars']

df_unhealthy=df.groupby(['Category'])[nutritional_content_1].mean().sort_values(by=nutritional_content_1,ascending=False)
print("\nAverage Nutritional Content of Unhealthy components by Category:")
df_unhealthy

```

```python
# Determine the average nutritional content of popular menu categories
nutritional_content_2=['Dietary Fiber','Protein','Vitamin A (% Daily Value)','Vitamin C (% Daily Value)','Calcium (% Daily Value)','Iron (% Daily Value)']
df_healthy=df.groupby(['Category'])[nutritional_content_2].mean().sort_values(by=nutritional_content_2,ascending=False)
print("\nAverage Nutritional Contents of Healthy components by Category:")
df_healthy
```

```python
# Find the rank of Average Nutritional Content by Category of Unhealthy:
df_unhealthy_rank=df_unhealthy.rank()
print("Ranking Categories based on UnHealthy Indivitual Component:")
df_unhealthy_rank
```

```python
# Find the rank of Average Nutritional Content by Category:
df_healthy_rank=df_healthy.rank()
print("Ranking Categories based on Healthy Indivitual Component:")
df_healthy_rank
```

```python
#Find the mean rank of unhealthy food by Category
df_unhealthy_rank_mean=df_unhealthy_rank.mean(axis=1)
print('\nRanking Categories based on Unhealthy Food:" ')
df_unhealthy_rank_mean
```

```python
#Find the mean rank ofhealthy food by Category
df_healthy_rank_mean=df_healthy_rank.mean(axis=1)
print('\n Ranking Categories based on Healthy Food:')
df_healthy_rank_mean
```

#### Analysis 

* On finding the average nutritional content of menu categories,intially i found the ranks of each component categorizedas Unhealthy and healthy.
* On analysing the data I found that Breakfast,Chicken & Fish,Beef & Pork almost have similarly ranked healthy and unhealthy components.
* Upon observing the data, we concluded that the McDonald's menu contains a relatively balanced mix of healthy and unhealthy nutritional components. For instance, in the Breakfast category, the rank is 7.83 in the unhealthy category and 6.33 in the healthy category.
* In terms of healthy menu options, salads stand out as the best choice
